```python
top10 = df['연수명'].value_counts(ascending=False)[:10]
top10
```

```python
top10_title = top10.index.tolist()
```

```python
top10_reviews = df[df['연수명'].isin(top10_title)]
```

```python
top10_reviews
```

```python
top10_reviews.info()
```

```python
top10_reviews_title = top10_reviews['연수명'].unique().tolist()
```

```python
top10_reviews_title
```

```python
avg_score = {}

for t in top10_reviews_title:
    avg = top10_reviews[top10_reviews['연수명'] == t]['별점'].mean()
    avg_score[t] = round(avg, 2)
```

```python
avg_score
```

```python
plt.figure(figsize=(10, 5))
plt.title('연수 평균 평점 (top 10: 리뷰 수)\n', fontsize=17)
plt.xlabel('연수 제목')
plt.ylabel('평균 평점')
plt.xticks(rotation=90)
plt.ylim([4.6, 5])

for x, y in avg_score.items():
    plt.bar(x, y, color='b')
    plt.text(x, y, '%.2f' % y, 
             horizontalalignment='center',  
             verticalalignment='bottom')    

plt.show()
```
