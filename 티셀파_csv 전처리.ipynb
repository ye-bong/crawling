```python
import pandas as pd
```

```python
df = pd.read_csv('t_data_2.csv', index_col=0) 
```

```python
df
```

```python
df.columns = ['연수명', '내용', '별점', '등록일']
```

```python
df
```

```python
df['별점'] = df['별점'].apply(lambda x: x.rstrip('점'))
```

```python
df['분류'] = None
```

```python
df['별점'] = df['별점'].astype(int)
```

```python
for i in range(len(df)):
    if (df.iloc[i, 2] == 5) or (df.iloc[i, 2] == 4):
        df.iloc[i, 4] = '긍정'
    else:
        df.iloc[i, 4] = '부정'
```

```python
df
```

```python
df.to_csv('t_data_2023_2022.csv')
```
