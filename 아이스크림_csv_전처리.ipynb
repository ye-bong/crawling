```python
import pandas as pd
```

```python
df_1 = pd.read_csv('데이터_원본/아이스크림_ICT_2.csv', index_col=0)  
```

```python
df_2 = pd.read_csv('데이터_원본/아이스크림_교과자료.csv') 
```

```python
df_2.columns = ['연수명', '내용', '별점', '등록일']
df_3.columns = ['연수명', '내용', '별점', '등록일']
df_4.columns = ['연수명', '내용', '별점', '등록일']
df_5.columns = ['연수명', '내용', '별점', '등록일']
```

```python
df = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7])
```

```python
df.iloc[1,2 ].lstrip('rate')
```

```python
for i in range(len(df)):
    df.iloc[i, 2] = df.iloc[i, 2].lstrip('rate')
```

```python
df['별점'].unique()
```

```python
df['분류'] = None
```

```python
for i in range(len(df)):
    if (df.iloc[i, 2] == '5') or (df.iloc[i, 2] == '4'):
        df.iloc[i, 4] = '긍정'
    else:
        df.iloc[i, 4] = '부정'
```

```python
df['분류'].unique()
```

```python
df['별점'] = df['별점'].astype(int)
```
